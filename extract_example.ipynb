```python
# import packages
import pytesseract
from pdf2image import convert_from_path
```

PDF file after 1990

```python
# Set path to input PDF file
pdf_path = 'Dataset/Afghanistan/2009_667703.pdf'

# Convert PDF to PIL images using pdf2image
images = convert_from_path(pdf_path, dpi=300)

pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'

# Loop over each image and extract text using Tesseract
text_list = []
for image in images:
    text = pytesseract.image_to_string(image)
    text_list.append(text)

# Combine text from all pages into a single string
text_all_pages = '\n'.join(text_list)

# Print and save extracted text
print(text_all_pages)
```

PDF file before 1990

```python
pdf_path_1990 = 'Afghanistan/1956_3865163.pdf'
# Convert PDF to PIL images using pdf2image
images_1990 = convert_from_path(pdf_path_1990, dpi=300)
# Loop over each image and extract text using Tesseract
text_list = []
for image in images_1990:
    text = pytesseract.image_to_string(image)
    text_list.append(text)

# Combine text from all pages into a single string
text_all_pages_1990 = '\n'.join(text_list)

# Print and save extracted text
print(text_all_pages_1990)
```
